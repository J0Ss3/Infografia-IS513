```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.patches as mpatches
```

```python
df=pd.read_csv("Infografia.csv")
```

```python
df
```

```python

df = df.sort_values(by='Popularidad', ascending=False)
plt.figure(figsize=(15,5))
plt.bar(df['Nombre'], df['Popularidad'])
plt.xlabel('Lenguaje de programacin')
plt.ylabel('Popularidad')
plt.title('Popularidad de los lenguajes de programacin')
plt.bar(df['Nombre'], df['Popularidad'], color=['blue', 'green', 'red', 'orange', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan', 'magenta', 'yellow', 'black', 'darkgreen', 'silver', 'gold', 'maroon', 'navy', 'teal', 'indigo'])
plt.show()
```

```python
df = df.sort_values(by='Oferta Laboral', ascending=False)
plt.figure(figsize=(15,5))
plt.bar(df['Nombre'], df['Oferta Laboral'])
plt.xlabel('Lenguaje de programacin')
plt.ylabel('Oferta Laboral')
plt.title('Cantidad de Oferta Laboral en cada lenguaje de programacion')
plt.bar(df['Nombre'], df['Oferta Laboral'], color=['blue', 'green', 'red', 'orange', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan', 'magenta', 'yellow', 'black', 'darkgreen', 'silver', 'gold', 'maroon', 'navy', 'teal', 'indigo'])
plt.show()
```

```python
df = df.sort_values(by='Interes', ascending=False)
plt.figure(figsize=(15,5))
plt.bar(df['Nombre'], df['Interes'])
plt.xlabel('Lenguaje de programacin')
plt.ylabel('Interes')
plt.title('Cantidad de Interes en cada lenguaje de programacion')
plt.bar(df['Nombre'], df['Interes'], color=['blue', 'green', 'red', 'orange', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan', 'magenta', 'yellow', 'black', 'darkgreen', 'silver', 'gold', 'maroon', 'navy', 'teal', 'indigo'])
plt.show()
```

```python

pastel_Ejecucion = df['Ejecucion'].value_counts()

plt.pie(pastel_Ejecucion , labels=pastel_Ejecucion .index, autopct='%1.1f%%')
plt.title('Frecuencia de tipos de ejecucin')
plt.show()
```

```python

pastel_Ejecucion = df['Tipado'].value_counts()

plt.pie(pastel_Ejecucion , labels=pastel_Ejecucion .index, autopct='%1.1f%%')
plt.title('Frecuencia de tipos de tipado')
plt.show()
```

```python
pastel_Paradigma = df['Paradigma'].value_counts()

plt.pie(pastel_Paradigma, labels=pastel_Paradigma .index, autopct='%1.1f%%')
plt.title('Frecuencia de tipos de paradigma')
plt.show()
```

```python

colors = {'Multiparadigma': 'blue', 'Orientado a Objetos': 'green', 'Estructurado': 'red', 'Declarativo': 'orange', 'Sistemas Operativos': 'purple', 'Gestion de Datos': 'brown', 'Imperativo': 'pink'}

plt.figure(figsize=(15,5))
plt.bar(df['Paradigma'], df['Oferta Laboral'],color=[colors[i] for i in df['Paradigma']])
plt.xlabel('Paradigma')
plt.ylabel('Oferta Laboral')
plt.title('Oferta Laboral por Paradigma')
plt.show()
```

```python
fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(111, projection='3d')

colors = ['blue', 'green', 'red', 'orange', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan', 'magenta', 'yellow', 'black', 'white', 'silver', 'gold', 'maroon', 'navy', 'teal', 'indigo']

patches = []
for i in range(len(df)):
    patches.append(mpatches.Patch(color=colors[i], label=df['Nombre'][i]))

plt.legend (handles=patches, bbox_to_anchor=(-0.2, 0.5, 0.2, 0.4))




for i in range(len(df)):
    ax.scatter(df['Interes'][i], df['Oferta Laboral'][i], df['Popularidad'][i], c=colors[i], s=100)

ax.set_xlabel('Inters')
ax.set_ylabel('Oferta Laboral')
ax.set_zlabel('Popularidad')
ax.set_title('Comparacin de Inters, Oferta Laboral y Popularidad por Lenguaje de Programacin')




plt.show()
```
